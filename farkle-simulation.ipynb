```python
import random

class Dice:
    def __init__(self, sides=6):
        self.sides = sides

    def roll(self):
        return random.randint(1, self.sides)

class DiceHand:
    def __init__(self, num_dice=6):
        self.dice = [Dice() for _ in range(num_dice)]

    def roll(self):
        rolls = [d.roll() for d in self.dice]
        self.dice = [d for d in self.dice if d.roll() not in rolls]
        return rolls

# Create a dice hand with 6 dice
hand = DiceHand()

# Roll the dice until there are no more dice left
while hand.dice:
    rolls = hand.roll()
    print(rolls)

print("No more dice to roll!")
        
```

```python
import random

def roll_dice():
    return random.randint(1, 6)

def play_game():
    dice = [roll_dice() for _ in range(6)]
    while len(dice) > 0:
        print(f"Current dice: {dice}")
        matching_dice = [die for i, die in enumerate(dice) if dice.count(die) > 1]
        if matching_dice:
            dice = [die for die in dice if die not in matching_dice]
        else:
            dice = [roll_dice() for _ in range(len(dice))]
    print("No dice left!")

play_game()
```

```python
import random

def roll_dice():
    return random.randint(1, 6)

def get_score():
    dice = [roll_dice() for _ in range(6)]
    score = 0
    for die in dice:
        if die == 1:
            score += 100
        elif die == 5:
            score += 50
    return score

score = get_score()
print(score)
```

```python
import random

class Dice:
    def __init__(self, sides=6):
        self.sides = sides

    def roll(self):
        return random.randint(1, self.sides)

class DiceHand:
    def __init__(self, num_dice=6):
        self.dice = [Dice() for _ in range(num_dice)]

    def roll(self):
        def sort_arr(arr: list):
            return sorted(arr)
        rolls = [d.roll() for d in self.dice]
        return sort_arr(rolls)

class Scoring:
    def __init__(self, arr) -> None:
        self.arr = arr

    def point_summary(self) -> int:
        points = 0
        hot_dice = 0

        for i in self.arr:
            if i == 1:
                points += 100
                hot_dice += 1
            elif i == 5:
                points += 50
                hot_dice += 1

        return points, hot_dice

player_score = 0
die_in_play = 6
while die_in_play > 0:
    hand = DiceHand(num_dice=die_in_play).roll()
    hand_score, hot_die = Scoring(arr=hand).point_summary()
    if hand_score > 0:
        status = "Continue"
        die_in_play -= hot_die
        player_score += hand_score
        print(f"{status}\t{player_score}\t{hand_score}\t{hand}")
    else: 
        status = "Farkle! "
        die_in_play -= hot_die
        player_score += hand_score
        print(f"{status}\t{player_score}\t{hand_score}\t{hand}")
        break
    
```
